```python
# Import libraries for working with data
import requests
from bs4 import BeautifulSoup
import pandas as pd
import re
```

```python
# recipes pulled based on ingredients or types
# chicken
# fish and seafood
# pasta and noodles
# beef
# pork
# baking
# dessert

# list_of_recipes = []                                                                            # list to store recipes
page_count = 1                                                                                    # start counting pages accessed
number_of_recipes = 1                                                                             # count number of recipes
while number_of_recipes < 30:                                                                     # looking for 100+ recipes
    print("--- Page number : " + str(page_count) + " ---")
    page_url = str("https://www.simplyrecipes.com/recipes/course/dessert/page/" + str(page_count) + "/")
    req = requests.get(page_url)                                                                  # request the HTML content
    my_soup = BeautifulSoup(req.text, 'html.parser')                                              # make soup using html parser
    if my_soup.find("ul", {"class" : "entry-list"}) is not None:                                  # check if page exists
        recipes = my_soup.find("ul", {"class" : "entry-list"}).find_all("a")                      # check for access to recipes
        for i in recipes:
            link = i.get("href")
            recipe_page = requests.get(link)
            recipe_soup = BeautifulSoup(recipe_page.text, 'html.parser')
            recipe_name = recipe_soup.find("h1", {"class" : "entry-title"})
            if recipe_name is not None:
                recipe_ingredients = recipe_soup.find("div", {"class" : "entry-details recipe-ingredients"})
                if recipe_ingredients is not None:
                    print(str(number_of_recipes) + str(link))
                    print(recipe_name.text)
                    print(recipe_ingredients.find_all("ul"))
                    number_of_recipes = number_of_recipes + 1
                    new_row = [link, recipe_name.text, recipe_ingredients]
                    list_of_recipes.append(new_row)
    page_count = page_count + 1
```
