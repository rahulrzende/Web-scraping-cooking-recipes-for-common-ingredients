```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
req = requests.get('https://www.simplyrecipes.com/recipes/ingredient/chicken/pinned/page/1/')   # request the HTML 
my_soup = BeautifulSoup(req.text, 'html.parser')                                                # make soup using html parser
```

```python
my_soup
```

```python
temp = pd.DataFrame(columns=['url','name','ingredient'])
list = []
recipes = my_soup.find("ul", {"class" : "entry-list"}).find_all("a")
count=1
for i in recipes:
    link = i.get("href")
    recipe_page = requests.get(link)
    recipe_soup = BeautifulSoup(recipe_page.text, 'html.parser')
    recipe_name = recipe_soup.find("h1", {"class" : "entry-title"})
    if recipe_name is not None:
        recipe_ingredients = recipe_soup.find("div", {"class" : "entry-details recipe-ingredients"}).find_all("ul")
        if recipe_ingredients is not None:
            print(link); 
            print(recipe_name.text)
            print(recipe_ingredients)
            new_dataframe = [link, recipe_name.text, recipe_ingredients]
            list.append(new_dataframe)
            count = count + 1
print(count)
```
